### Step 6

```python
# Create a spending flag for each user based on the average price across all their orders
# If the mean of the prices of products purchased by a user is lower than 10, 
# then flag them as a “Low spender.”
# If the mean of the prices of products purchased by a user is higher than or equal to 10, 
# then flag them as a “High spender.”

# Create an "avg_order” column
ords_prods_merge['avg_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.mean)
ords_prods_merge.head()
```

```python
# Create a spending flag
ords_prods_merge.loc[ords_prods_merge['avg_order'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_order'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.head()
```

### Step 7

```python
# Create an order frequency flag that marks the regularity of a user’s ordering behavior 
# according to the median in the “days_since_prior_order” column. 

# Create an "median_dspo” column
ords_prods_merge['median_dspo'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
ords_prods_merge.head()
```

```python
# Create an order frequency flag
# If the median of “days_since_prior_order” is higher than 20, 
# then the customer should be labeled a “Non-frequent customer.”
ords_prods_merge.loc[ords_prods_merge['median_dspo'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
# If the median is higher than 10 and lower than or equal to 20, 
# then the customer should be labeled a “Regular customer.”
ords_prods_merge.loc[(ords_prods_merge['median_dspo'] > 10) & (ords_prods_merge['median_dspo'] <= 20), 'order_freq_flag'] = 'Regular customer'
```

```python
# If the median is lower than or equal to 10, 
# then the customer should be labeled a “Frequent customer.”
ords_prods_merge.loc[ords_prods_merge['median_dspo'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['order_freq_flag'].value_counts(dropna = False)
```
